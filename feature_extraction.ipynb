```python
#test_images[0]
for row in range(128):
        for col in range(128):
            if test_images[21][row][col] == 255.:
                test_images[21][row][col] = 255.
            else:
                test_images[21][row][col] = 0.

plt.imshow(test_images[21])
plt.show()
```

```python
X_image = test_images[21].copy()
coordinates = []
for x in range(128):
    for y in range(128):
        if X_image[x][y] == 255.:
            coordinates.append([x,y])

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(coordinates)
y_kmeans = kmeans.predict(coordinates)
#kmeans.labels_
number_one = []
number_two = []
number_three = []
for label in range(len(y_kmeans)):
    if y_kmeans[label] == 0.:
        number_one.append(coordinates[label])
    elif y_kmeans[label] == 1.:
        number_two.append(coordinates[label])
    else:
        number_three.append(coordinates[label])
number_one
```

```python
row = []
col = []
for point in number_three:
    row.append(point[0])
    col.append(point[1])
row = np.array(row)
col = np.array(col)
for old_image in [number_one, number_two, number_three]:
    new_image = np.zeros((128,128))
    for i in range(128):
        for j in range(128):
            if [i,j] in old_image:
                new_image[i][j] = 255.
    plt.imshow(new_image)
    plt.show()
```
